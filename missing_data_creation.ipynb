```python
import pandas as pd
import numpy as np
df = pd.read_csv('PhiUSIIL_Phishing_URL_Dataset.csv')
```

```python
np.random.seed(16) 

# We generate random numbers between 0 and 1. If < 0.05, it becomes True.
mask = np.random.random(df.shape) < 0.05

# Apply the mask to the dataframe
df_missing = df.mask(mask)
```

```python
# Check the actual percentage of missing data
total_cells = np.prod(df_missing.shape)
missing_cells = df_missing.isnull().sum().sum()
percent = (missing_cells / total_cells) * 100
print(f"Percentage of missing data: {percent:.2f}%")
```

```python
df_missing.to_csv('Phishing_URL_ds_5percent_missing.csv', index=False)
```
